## Prepare the features

```python
# create the features
df = get_features_lgbm(data)
features = get_features_from_file("features_lgbm.json")

# load the model
#pipeline_lgbm   = joblib.load(os.path.join(ROOT_DIR, "models", "pipeline_lgbm.pkl"))
```

## Training

```python
from sklearn.multioutput import MultiOutputRegressor
from lightgbm import LGBMRegressor

#prepare the hierharchy
nationalities_cols = [
    "tickets_saudi",
    "tickets_egyptian",
    "tickets_pakistani",
    "tickets_indian",
    "tickets_yemeni",
    "tickets_indonesian",
    "tickets_jordanian",
    "tickets_USA", 
    "tickets_britain"]
other_tag = 'tickets_other_nationalities'

#prepare the dataframe
df[other_tag] = df['passengers'] - df[nationalities_cols].sum(axis=1)
#df_long = df.melt(id_vars='train_departure_date_short', value_vars=nationalities_cols+other_tag, var_name='nationality', value_name='Y')
#df_long = df_long.rename(columns={'train_departure_date_short': 'ds'})

#take training data
n = df.shape[0]
train_size = int(n*0.8)
test_size = n - train_size

X = df[features]
y = df[nationalities_cols + [other_tag]]

X_train, X_test, y_train, y_test = X[:train_size], X[train_size:], y[:train_size], y[train_size:]

#fit
lgbm_params = {
    "objective": "regression_l2",        # best fit regression_l2
    "metric": "rmse",                 # rmse
    #"learning_rate": 0.1,
    "num_leaves": 3, # reduced to 3
    #"max_depth": 3,                  
    "min_data_in_leaf": 10, 
    "feature_fraction": 1, #changed to 1
    "bagging_fraction": 1, #idem
    #"bagging_freq": 20,
    #"lambda_l1": 0.8,
    "lambda_l2": 0.8,
    "verbosity": -1,
    "boosting": "gbdt",                # gbdt
    "n_estimators": 100, #down to 100
    #"subsample":0.8,
    #"colsample_bytree":0.8,
    "random_state":42
}

model = MultiOutputRegressor(LGBMRegressor(**lgbm_params))
fitted_model = model.fit(X_train, y_train)
y_pred = fitted_model.predict(X_test)
```

## Evaluation

```python
y_pred
```

```python
def evaluation(model, X_test, y_test):
    # forecasting
    y_pred = model.predict(X_test)

    # funciÃ³n interna para evaluar y mostrar mÃ©tricas
    def eval_metrics(true, pred, label=""):
        r2 = r2_score(true, pred)
        mae = mean_absolute_error(true, pred)
        mse = mean_squared_error(true, pred)
        print(f"{label}RÂ²: {r2:.4f} | MAE: {mae:.2f} | MSE: {mse:.2f}")

    print("\nEvaluating output by column")
    for i in range(y_test.shape[1]):
        eval_metrics(y_test.iloc[:, i], y_pred[:, i], label=f"Column {y_test.columns[i]}: ")

    # total
    y_test_total = y_test.sum(axis=1)
    y_pred_total = y_pred.sum(axis=1)
    print("\n--- Total (Sum of all columns) ---")
    eval_metrics(y_test_total, y_pred_total, label="Total: ")

    return y_pred

y_pred = evaluation(fitted_model, X_test, y_test)
```
